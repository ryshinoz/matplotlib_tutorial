## plot

```python
X = np.linspace(-5, 5, 10)
Y = X * 2

plt.plot(X, Y)
```

線の色、太さ、スタイルなども変更できます。
設定できるプロパティは[lines.Line2D](https://matplotlib.org/api/lines_api.html#matplotlib.lines.Line2D)を参照してください。

```python
plt.plot(X, Y, color="red", linewidth=5.0, linestyle=':')
```

## legend

```python
plt.plot(X, Y, label=r'$ y=2 x $')

plt.legend()
```

locオプションで表示位置を設定できます。詳細は[legend and legend_handler](https://matplotlib.org/api/legend_api.html)を参照してください。

```python
plt.plot(X, Y, label=r'$ y=2 x $')

plt.legend(loc='right')
```

## xlim, ylim, xticks, yticks

```python
plt.plot(X, Y, label=r'$ y=2 x $')

plt.legend(loc='right')

plt.xlim(-10, 10)
plt.ylim(-20, 20)
```

```python
plt.plot(X, Y, label=r'$ y=2 x $')

plt.legend(loc='right')

plt.xticks([-5, 0, 5])
plt.yticks([-10, 0, 10])
```

```python
plt.plot(X, Y, label=r'$ y=2 x $')

plt.legend(loc='right')

plt.xticks([-5, 0, 5], ['a', 'b', 'c'])
plt.yticks([-10, 0, 10], ['A', 'B', 'C'])
```
